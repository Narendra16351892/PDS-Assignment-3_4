```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the diabetes dataset
data = pd.read_csv("diabetes.csv")

# Set seed for reproducibility
np.random.seed(25)

# Take a random sample of 25 observations
sample = data.sample(n=25)

# Calculate the 98th percentile of BMI for the sample and population
sample_98th_percentile_bmi = np.percentile(sample['BMI'], 98)
population_98th_percentile_bmi = np.percentile(data['BMI'], 98)

# Create a line graph to compare the 98th percentile of BMI
plt.figure(figsize=(8, 6))

# Plot for sample 98th percentile BMI
plt.plot([0, 1], [sample_98th_percentile_bmi, sample_98th_percentile_bmi], marker='o', linestyle='-', color='blue', label='Sample')

# Plot for population 98th percentile BMI
plt.plot([0, 1], [population_98th_percentile_bmi, population_98th_percentile_bmi], marker='o', linestyle='-', color='orange', label='Population')

# Add labels and title
plt.xticks([0, 1], ['Sample', 'Population'])
plt.title('98th Percentile BMI Comparison')
plt.ylabel('BMI')
plt.legend()

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the diabetes dataset
df = pd.read_csv('diabetes.csv')

# Extract the BloodPressure column
blood_pressure = df['BloodPressure']

# Create histogram
plt.figure(figsize=(10, 6))
plt.hist(blood_pressure, bins=20, color='skyblue', edgecolor='black', alpha=0.7)
plt.title('Blood Pressure Distribution')
plt.xlabel('Blood Pressure')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
